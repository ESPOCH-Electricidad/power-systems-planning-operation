# Catálogo de casos

Notebook de exploración de datos para la asignatura.

```
from pathlib import Path
import yaml, pandas as pd
base=Path('..')
rows=[]
for meta in base.glob('*/metadata.yaml'):
    with open(meta, encoding='utf-8') as f:
        d=yaml.safe_load(f)
    d['carpeta']=meta.parent.name
    rows.append(d)
pd.DataFrame(rows)
```
