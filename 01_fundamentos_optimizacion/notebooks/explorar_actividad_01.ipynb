# Exploración de datos — Actividad 01

Notebook de exploración de datos para la asignatura.

```
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
base=Path('..')/'actividades'
df=pd.read_csv(base/'datos_actividad_01.csv')
print(df)
g=df[df['tipo'].eq('generador')].copy()
for c in ['costo_usd_mwh','pmin_mw','pmax_mw','emisiones_tco2_mwh']:
    g[c]=pd.to_numeric(g[c], errors='coerce')
g.plot(x='id', y=['costo_usd_mwh','pmax_mw'], kind='bar')
plt.title('Costo variable y capacidad máxima')
plt.tight_layout(); plt.show()
```
