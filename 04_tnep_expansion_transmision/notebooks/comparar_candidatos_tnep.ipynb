```
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
base=Path('..')/'actividades'
cor=pd.read_csv(base/'datos_actividad_04_corredores.csv')
print(cor)
cor.plot(x='corredor', y=['costo_kusd','capacidad_mw'], kind='bar', figsize=(10,4))
plt.title('Costo y capacidad por corredor')
plt.tight_layout(); plt.show()
```
