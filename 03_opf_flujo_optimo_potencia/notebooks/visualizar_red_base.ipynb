# Visualización de red — OPF

Notebook de exploración de datos para la asignatura.

```
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
base=Path('..')/'actividades'
lin=pd.read_csv(base/'datos_actividad_03_lineas.csv')
pos={1:(0,1),2:(1,1.5),3:(1,0.5),4:(2,1),5:(3,0.8)}
fig,ax=plt.subplots(figsize=(7,4))
for _,r in lin.iterrows():
    x1,y1=pos[int(r['desde'])]; x2,y2=pos[int(r['hasta'])]
    ax.plot([x1,x2],[y1,y2], linewidth=max(1,r['limite_mw']/50))
    ax.text((x1+x2)/2,(y1+y2)/2,r['linea'])
for b,(x,y) in pos.items():
    ax.scatter(x,y,s=500); ax.text(x,y,str(b),ha='center',va='center',color='white',fontweight='bold')
ax.set_axis_off(); ax.set_title('Red didáctica OPF'); plt.show()
```
